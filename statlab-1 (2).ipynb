## Основные теоретические положения
Случайный вектор $\overrightarrow{\xi} = \left(\xi_1, \dots, \xi_n\right)^\intercal:\Omega\to \mathbb{R}^n$ называется нормальным (имеет многомерное нормальное распределение), если существует вектор независимых стандартных нормальных случайных величин $\overrightarrow{\eta} = \left(\eta_1, \dots, \eta_m\right)^\intercal$, матрица A размерности $n \times m$ и вещественный вектор $\overrightarrow{b} = (b_1,\dots,b_n)^\intercal$, такие что: 

 $\overrightarrow{\xi} = A\overrightarrow{\eta} + \overrightarrow{b} $.

Плотность многомерного невырожденного нормального распределения имеет вид: $p_{\overrightarrow{\xi}}\left(x_1, \dots, x_n\right) = \frac{1}{(2\pi)^{n/2}\sqrt{\det \Sigma}} \exp\left(-\frac{1}{2}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)^\intercal \Sigma^{-1}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)\right)$

где $\overrightarrow{\mu} = \mathbb{E}\overrightarrow{\xi} = \left(\mathbb{E}\xi_1, \dots, \mathbb{E}\xi_n\right)^\intercal$, $\Sigma = \left\{\sigma_{i,j}\right\}_{i,j=1}^n$, где $\sigma_{i,j} = \mathop{\mathrm{cov}}\left(\xi_i,\xi_j\right)$, $\overrightarrow{x} = (x_1, \dots, x_n)^\intercal$

Свойства многомерного нормального распределения:
1. Линейная комбинация случайных переменных, принадлежащих многомерному нормальному распределению, также является нормально распределенной.
2. Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормально распределенными.
<br>$\mathcal{p}_{\xi|\eta=y}(x) = \frac{\mathcal{p}_{\xi,\eta}(x,y)}{\mathcal{p}_{\eta}(y)}$
3. Параметры многомерного нормального распределения полностью описывают его форму.
4. Многомерное нормальное распределение устойчиво относительно линейных преобразований. Если $\overrightarrow{\xi} \sim \mathcal{N}\left(\overrightarrow{\mu}, \Sigma\right)$, $A$ - произвольная матрица размерности $m \times n$, $\overrightarrow{b}$ - произвольный вектор размерности $m \times 1$, то $A\overrightarrow{\xi}+\overrightarrow{b} \sim \mathcal{N}\left(A\overrightarrow{\mu}+\overrightarrow{b}, A \Sigma A^\intercal\right)$.

#### 1) Вычислим вектор математических ожиданий и ковариационные характеристики:
Преобразуем квадратичную форму в показателе экспоненты
$\sqsupset q(x,y) = 4x^2+9xy+16y^2+7x-14y = (4x^2+9xy+7x) + 16y^2-14y+14 =\newline
4\bigl(x^2+2x\left(\frac{9}{8}y+\frac{7}{8}\right)\bigr)+16y^2-14y+14=\newline
4\bigl(x^2+2x\left(\frac{9}{8}y+\frac{7}{8}\right)+\left(\frac{9}{8}y+\frac{7}{8}\right)^2-\left(\frac{9}{8}y+\frac{7}{8}\right)^2\bigr)+16y^2-14y+14=\newline

Откуда имеем:
$ \mathrm{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} -2 \\ 1 \end{pmatrix} \newline
\Sigma^{-1} = \begin{pmatrix}
4 & \frac{9}{2} \\
\frac{9}{2} & 16\end{pmatrix}
,\,\mathrm{det}(\Sigma^{-1})=\frac{175}{4},\,
\Sigma=\frac{4}{175}\begin{pmatrix}
16 & -\frac{9}{2} \\
-\frac{9}{2} & 4\end{pmatrix}\newline
\mathrm{D}\xi=\frac{64}{175},\,\mathrm{D}\eta=\frac{16}{175}\newline
\mathrm{Cov}(\xi,\eta)=-\frac{18}{175}\newline
\mathrm{\rho}=\frac{\mathrm{Cov}(\xi,\eta)}{\sqrt{\mathrm{D}\xi\mathrm{D}\eta}}=\frac{-\frac{18}{175}}{\sqrt{\frac{32^2}{175^2}}}=-\frac{9}{16}\newline
C=\frac{1}{2\pi^{\frac{n}{2}}\sqrt{\mathrm{det}(\Sigma)}}=\frac{1}{2\pi\sqrt{\frac{4}{175}}}=\frac{\sqrt{175}}{4\pi}=\frac{5\sqrt{7}}{4\pi}\newline$
Тогда функция плотности имеет следующий вид:$
\newline
p_\xi,_\eta(x,y) = \frac{5\sqrt{7}}{4\pi}\mathrm{exp}\Biggl[-\frac{1}{2}
\Biggl(
\begin{pmatrix}
x \\ y
\end{pmatrix}
-
\begin{pmatrix}
-2 \\ 1
\end{pmatrix}

\begin{pmatrix}

\Biggr]
$
Построим график плотности распределения:

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator

C = (5 * math.sqrt(7)) / (4 * math.pi)

Sigma = np.array([[4, 4.5], [4.5, 16]])

mu = [-2, 1]


def p(x, y):
    x_mu = x - mu[0]
    y_mu = y - mu[1]
    a = np.array([x_mu, y_mu])
    a_t = np.transpose(a)
    return math.exp(-0.5 * np.dot(np.dot(a_t, Sigma), a))


fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

x_cnt = 250
y_cnt = 250
X = np.linspace(mu[0] - 1.5, mu[0] + 1.5, x_cnt)
Y = np.linspace(mu[1] - 1.5, mu[1] + 1.5, y_cnt)
Z = np.array([[0] * x_cnt for _ in range(y_cnt)], dtype=float)
for x_i in range(x_cnt):
    for y_i in range(y_cnt):
        v = np.array([X[x_i], Y[y_i]], dtype=float)
        Z[y_i][x_i] = p(v[0], v[1])

X, Y = np.meshgrid(X, Y)

surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.set_zlim(0, 1)
ax.zaxis.set_major_locator(LinearLocator(10))
plt.show()
```
